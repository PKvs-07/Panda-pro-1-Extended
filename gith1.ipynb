```python
import pandas as pd
```

```python
data={"Name": ["Alice","Bob","Charlie","David","Emma"],
      "Age":[24,27,22,32,29],
"Department": ["HR","Finance","IT","Marketing","Sales"],
"Salary" : [45000, 54000,50000,61000,58000]}       
             
```

```python
df= pd.DataFrame(data)
print (df)
```

Average age of employees

```python
age=df["Age"].mean()
print (age)
```

Total salary exepnse

```python
salary=df["Salary"].sum()
print (salary)
```

Maximum and minimum age 

```python
min=df["Age"].min()
max=df["Age"].max()
print (min)
print(max)
```

Salary difference from company average

```python
df["Avg"] = df["Salary"]- df["Salary"].mean()
print(df["Avg"])
```

```python
Add a 10% annual bonus column
```

```python
df["Bonus"]=df["Salary"]* 0.10
print (df)
```

Names and salary of emplloyees younger than 30 

```python
age=df[df["Age"] < 30]
print (age)
```

Employees with salary between 45,000 and 60,000

```python
mid_salary=df[(df["Salary"] >= 45000) & (df["Salary"] <= 60000)]
print (mid_salary)
```

Names and salries of employees younger than 30

```python
young=df[df["Age"] < 30] [["Name","Salary"]]
print (young)
```

Employees in IT deartment

```python
it=df[df["Department"] == "IT"]
print (it)
```

Employees with the highest salary

```python
high=df.loc[df["Salary"].idxmax()]
print ("Employee with the highest salary:\n", high)
```

Number of employees per department

```python
dep=df["Department"].value_counts()
print(dep)
```

Medain salary

```python
median=df["Salary"].median()
print("Medain Salary:", median)
```

Create a new coloumn for salary category (High/Low)

```python
df["Salary_Level"]=df["Salary"].apply(lambda x: "High" if x>55000 else "Low")
print (df)
```
